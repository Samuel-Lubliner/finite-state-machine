```python
# From Colman Lecture:

# Mealy finite state machine is a model of computation which describes how an output is obtained given an input and a current state
# This abstract machine can be in one of a finite number of states. The state the machine is in at any given time is called the current state.
# It can change from one state to another. This change is called a transition. 

# A fsm M = (S,I,O,f,g,s_0) consists of:
# Finite set S of states
# Finite set I of inputs
# Finite set O of outputs
# Transition function f : S X I -> S that assigns to each state and input pair a new state
# Output function g : S X I -> O that assigns to each state and input pair an output
# An initial state s_0

# A vending machine accepts the following coins: 5 cents, 10 cents, and 25 cents. When a total of 30 cents or more has been deposited, the machine immediately returns the amount in excess of 30 cents. When 30 cents has been deposited and after any excess is refunded, the customer can push a red button and receive an apple or push a yellow button and receive a banana. The machine resets afterwards.

# https://doc.sagemath.org/html/en/reference/combinat/sage/combinat/finite_state_machine.html
# `Transducer` creates a transducer, which is a finite state machine, whose transitions have input and output labels.

```

```python
transitions = {
    's0': [('s1', 5, 'n'), ('s2', 10, 'n'), ('s5', 25, 'n'), ('s0', 'R', 'n'), ('s0', 'Y', 'n')],
    's1': [('s2', 5, 'n'), ('s3', 10, 'n'), ('s6', 25, 'n'), ('s1', 'R', 'n'), ('s1', 'Y', 'n')],
    's2': [('s3', 5, 'n'), ('s4', 10, 'n'), ('s6', 25, 5), ('s2', 'R', 'n'), ('s2', 'Y', 'n')],
    's3': [('s4', 5, 'n'), ('s5', 10, 'n'), ('s6', 25, 10), ('s3', 'R', 'n'), ('s3', 'Y', 'n')],
    's4': [('s5', 5, 'n'), ('s6', 10, 'n'), ('s6', 25, 15), ('s4', 'R', 'n'), ('s4', 'Y', 'n')],
    's5': [('s6', 5, 'n'), ('s6', 10, 5), ('s6', 25, 20), ('s5', 'R', 'n'), ('s5', 'Y', 'n')],
    's6': [('s6', 5, 5), ('s6', 10, 10), ('s6', 25, 25), ('s0', 'R', 'apple'), ('s0', 'Y', 'banana')]
}

vending_machine = Transducer(transitions, initial_states=['s0'], final_states=['s6'])

vending_machine.graph().plot(edge_labels=True, figsize=30)
```
